# Sonar Rock vs Mine Prediction using Logistic Regression

This project is a simple machine learning project that predicts whether a sonar signal corresponds to a rock or a mine using a Logistic Regression model. The project utilizes the Sonar Mines vs Rocks dataset and demonstrates the basic steps involved in building and evaluating the model.

````python
# Importing THe dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
# Reading CSV FIle
# Data collection and Processing
sonar_data = pd.read_csv("./RockVsMine_Data.csv", header=None)
sonar_data.head()
````

````python
# NUmber of rows and columns
sonar_data.shape
sonar_data.describe() # describes the statistical method of the data
````

````python
sonar_data.dtypes.value_counts()
````

````python
# How many rocks and how many mines
sonar_data[60].value_counts()
````

````python
#Grouping data based on mine and rock
sonar_data.groupby(60).mean()
````

````python
#Seprate the data and the Label
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
# print(X)
# print(Y)
````

````python
# Spliting above data into training and test data

X_Train, X_Test, Y_Train, Y_Test = train_test_split(X,Y, test_size= 0.1, stratify= Y, random_state=1)
# print(X.shape, X_Train.shape, X_Test.shape)
````

````python
print(X_Train)
print(Y_Train)
````

````python
#Training model
# Logistic regression model

model = LogisticRegression()
````

````python
#Training
model.fit(X_Train, Y_Train)
````

````python
# Model evaluation
# Accuracy on training data

X_Train_Prediction = model.predict(X_Train)
training_data_accuracy = accuracy_score(X_Train_Prediction, Y_Train)
````

````python
print("Accuracy on training data is: ",training_data_accuracy)
````

````python
# Model evaluation
# Accuracy on testing data

X_Test_Prediction = model.predict(X_Test)
test_data_accuracy = accuracy_score(X_Test_Prediction, Y_Test)
````

````python
print("Accuracy score for test data is: ", test_data_accuracy)
````
